```python
import pandas as pd
import missingno as msno
```

```python
#Insert original Data
data_mtk = pd.read_csv('../DataSource/MASTER_MTK.csv')
data_biologi = pd.read_csv('../DataSource/MASTER_BIOLOGI.csv')
data_fisika = pd.read_csv('../DataSource/MASTER_FISIKA.csv')
data_if = pd.read_csv('../DataSource/MASTER_IF.csv')

data_akutansi = pd.read_csv('../DataSource/MASTER_AKUTANSI.csv')
data_ekonomi = pd.read_csv('../DataSource/MASTER_EKONOMI.csv')
data_hi = pd.read_csv('../DataSource/MASTER_HI.csv')
data_hukum = pd.read_csv('../DataSource/MASTER_HUKUM.csv')
data_komunikasi = pd.read_csv('../DataSource/MASTER_KOMUNIKASI.csv')

ipk_mtk = pd.read_csv('../DataSource/IPK_MTK.csv')
ipk_biologi = pd.read_csv('../DataSource/IPK_BIO.csv')
ipk_fisika = pd.read_csv('../DataSource/IPK_FISIKA.csv')
ipk_if = pd.read_csv('../DataSource/IPK_IF.csv')

ipk_akutansi = pd.read_csv('../DataSource/IPK_AKUTANSI.csv')
ipk_ekonomi = pd.read_csv('../DataSource/IPK_EKONOMI.csv')
ipk_hi = pd.read_csv('../DataSource/IPK_HI.csv')
ipk_komunikasi = pd.read_csv('../DataSource/IPK_KOMUNIKASI.csv')
ipk_hukum = pd.read_csv('../DataSource/IPK_HUKUM.csv')

```

# Data Matematika

```python
# msno.matrix(data_mtk)
# msno.matrix(ipk_mtk)
datafull_mtk = data_mtk.merge(ipk_mtk,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_mtk)
msno.bar(datafull_mtk)
```

# # DATA BIOLOGI

```python

# msno.matrix(data_biologi)
# msno.matrix(ipk_biologi)
datafull_bio = data_biologi.merge(ipk_biologi,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_bio)
msno.bar(datafull_bio)
```

# Data Fisika

```python
# msno.matrix(data_fisika)
# msno.matrix(ipk_fisika)
datafull_fisika = data_fisika.merge(ipk_fisika,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_fisika)
msno.bar(datafull_fisika)
```

# Data Informatika

```python
# msno.matrix(data_if)
# msno.matrix(ipk_if)
datafull_if = data_if.merge(ipk_if,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_if)
msno.bar(datafull_if)
```

# Data Akutansi

```python
# msno.matrix(data_akutansi)
# msno.matrix(ipk_akutansi)
datafull_akutansi = data_akutansi.merge(ipk_akutansi,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_akutansi)
msno.bar(datafull_akutansi)
```

# Data Ekonomi

```python
# msno.matrix(data_ekonomi)
# msno.matrix(ipk_ekonomi)
datafull_ekonomi = data_ekonomi.merge(ipk_ekonomi,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_ekonomi)
msno.bar(datafull_ekonomi)
```

# Data HI

```python
# msno.matrix(data_hi)
# msno.matrix(ipk_hi)
datafull_hi = data_hi.merge(ipk_hi,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_hi)
msno.bar(datafull_ekonomi)
```

# Data Komunikasi

```python
# msno.matrix(data_komunikasi)
# msno.matrix(ipk_komunikasi)
datafull_komunikasi = data_komunikasi.merge(ipk_komunikasi,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_komunikasi)
msno.bar(datafull_komunikasi)
```

# Data Hukum

```python
# msno.matrix(data_hukum)
# msno.matrix(ipk_hukum)
datafull_hukum = data_hukum.merge(ipk_hukum,left_on='NIM', right_on='NIM', how='left')
msno.matrix(datafull_hukum)
msno.bar(datafull_hukum)
```

# Data ipa/science

```python
data_mipa =pd.concat([datafull_fisika, datafull_mtk,datafull_bio,datafull_if])
data_mipa = data_mipa.reset_index(drop=True)
msno.matrix(data_mipa)
msno.bar(data_mipa)
```

# Data Social

```python
data_sosial =pd.concat([datafull_akutansi, datafull_ekonomi,datafull_hi,datafull_komunikasi, datafull_hukum])
data_mipa = data_mipa.reset_index(drop=True)
msno.matrix(data_sosial)
msno.bar(data_sosial)
```
